```python
#Importando Pandas, OS para leitura do arquivo e Matplotlib

import pandas as pd
import os
import matplotlib.pyplot as plt

os.chdir('D:\\Users\\Diogo\\Documents\\Pós Graduação\\Módulos\\TCC\\DataSets\\Futebol Brasileiro')
```

```python
#Importando o data set dos dados das partidas dos campeonatos brasileiros e os dados dos cartões amarelos

df_jogos=pd.read_csv('campeonato-brasileiro-full.csv',sep=',')
df_cartoes=pd.read_csv('campeonato-brasileiro-cartoes.csv',sep=',')
```

```python
#Verificando os dados do dataframe df_cartoes
df_cartoes.head()
```

```python
#Agrupando a quantidade de cartões por partida_id, clube, cartao(tipo de cartão)

df_cartoes_tratados = df_cartoes.groupby(['partida_id','clube','cartao']).size().reset_index()
df_cartoes_tratados.head(20)

#Renomeando a coluna de quantidade de cartões
df_cartoes_tratados.rename(columns={0:'quantidade_cartoes'},inplace=True)
df_cartoes_tratados.head()

#Identificando se o clube é o mandante ou o visitante
#Atribuindo os dados para um Data Frame Temporário
df_cartoes_tratados_temp = df_cartoes_tratados

#Criando a coluna Mandante no DataFrame temporário e atribuindo um valor aleatório
df_cartoes_tratados_temp['mandante'] = 3
```

```python
#Verificando as dimensões do Dataframe
df_cartoes_tratados_temp.shape
```

```python
#Exibindo os dados do Data Frame df_cartoes

df_cartoes_tratados_temp[(df_cartoes_tratados_temp.mandante != 1) & (df_cartoes_tratados_temp.mandante != 0)]
```

```python
#Percorrendo o Data Frame dos cartões
for linha in range(len(df_cartoes_tratados_temp)):

    #Coletando o Id 
    id=df_cartoes_tratados_temp.iloc[linha,0].astype(str)

    #Buscando o mandante do confronto pelo ID no Data Frame dos jogos
    mandante_jogos = df_jogos[['ID','mandante']].query('ID=='+id)
    mandante_jogos = mandante_jogos.iloc[0,1]

    #Coletando o time no Data Frame dos cartões
    mandante_cartoes = df_cartoes_tratados_temp.iloc[linha,1]
    
    #Identificando se o clube é mandante, caso seja o mesmo clube significa que o clube é mandante caso contrário o clube é
    #visitante
    
    if mandante_cartoes==mandante_jogos:
        #Atribui o valor 1 na Coluna Mandante
        df_cartoes_tratados_temp['mandante'].iloc[linha] = 1
        
    else:
        df_cartoes_tratados_temp['mandante'].iloc[linha] = 0
        
```

```python
#Alterando o nome da coluna partida_id para ID do DataFrame de cartões para que fique com o mesmo nome do DataFrame de 
# Jogos
df_cartoes_tratados_temp.rename(columns={'partida_id':'ID'},inplace=True)
df_cartoes_tratados_temp.head()
```

```python
# Conferindo algumas partidas no Data Frame de Jogos e identificando se o mandante foi atribuido corretamente.

df_jogos[df_jogos.ID==4608]
```

```python
#Filtrando somente os registros de cartões amarelos
df_cartoes_amarelos= df_cartoes_tratados_temp[(df_cartoes_tratados_temp.cartao == 'Amarelo')]
```

```python
df_cartoes_amarelos['ID'].value_counts()
```

```python
#Localizando os clubes dos confrontos e adicionando o cartão amarelo zerado no Dataframe de Cartoes tratados

for linha in range(len(df_cartoes_amarelos)):

        #Coletando o id da partida
        idStr=df_cartoes_amarelos.iloc[linha,0].astype(str)
        id=df_cartoes_amarelos.iloc[linha,0]

        #Coletando o dado de mandante
        mandante=df_cartoes_amarelos.iloc[linha,4]

        #Localizando a partida com o id coletado no DataFrame de Jogos
        clube = df_jogos[['ID','mandante','visitante']].query('ID=='+idStr )


        #Se o valor de mandante for verdadeiro o clube que será coletado será o visitante e vice-versa
        if mandante==1:
            clube=clube.iloc[0,2]

            #Criando a linha para incluir no Data Frame de cartões temporário
            nova_linha = {'ID':id,'clube':clube,'cartao':'Amarelo','quantidade_cartoes':0,'mandante':0}
        else:
            clube=clube.iloc[0,1]
            nova_linha = {'ID':id,'clube':clube,'cartao':'Amarelo','quantidade_cartoes':0,'mandante':1}

        #Incluindo a linha no Data Frame    
        df_cartoes_tratados_temp = df_cartoes_tratados_temp.append(nova_linha,ignore_index=True)    
#mandante_jogos.iloc[0,1]
df_cartoes_amarelos['cartao'].value_counts()
```

```python
#Conferindo se os registros dos clubes foram inseridos no DataFrame de Cartões
#df_cartoes_tratados_temp

#Ordenando o DataFrame de cartões temporário pelo ID
df_cartoes_tratados_temp=df_cartoes_tratados_temp.sort_values(by = ['ID'])

#Conferindo se todos os registros de cartões amarelos estão com todas as linhas
df_cartoes_tratados_temp['ID'].value_counts()

#Identificados os IDs e verificados que essas partidas tiveram somente cartões vermelhos. Os cartões vermelhos serão tratados
#posteriormente
#5615,6542,4929,6513
df_cartoes_tratados_temp
```

```python
#Agora iremos tratar os cartões vermelhos para que os clubes que não aparecem com valor zerado na partida seja incluidos
#no DataFrame de Cartões

#Filtrando somente os registros de cartões amarelos
df_cartoes_vermelhos= df_cartoes_tratados_temp[(df_cartoes_tratados_temp.cartao == 'Vermelho')]

#Identificando os registros que possuem somente um registro de cartão amarelo
df_cartoes_vermelhos=df_cartoes_vermelhos.drop_duplicates('ID',keep=False)

```

```python
#Localizando os clubes dos confrontos e adicionando o cartão amarelo zerado no Dataframe de Cartoes tratados

for linha in range(len(df_cartoes_vermelhos)):

        #Coletando o id da partida
        idStr=df_cartoes_vermelhos.iloc[linha,0].astype(str)
        id=df_cartoes_vermelhos.iloc[linha,0]

        #Coletando o dado de mandante
        mandante=df_cartoes_vermelhos.iloc[linha,4]

        #Localizando a partida com o id coletado no DataFrame de Jogos
        clube = df_jogos[['ID','mandante','visitante']].query('ID=='+idStr )


        #Se o valor de mandante for verdadeiro o clube que será coletado será o visitante e vice-versa
        if mandante==1:
            clube=clube.iloc[0,2]

            #Criando a linha para incluir no Data Frame de cartões temporário
            nova_linha = {'ID':id,'clube':clube,'cartao':'Vermelho','quantidade_cartoes':0,'mandante':0}
        else:
            clube=clube.iloc[0,1]
            nova_linha = {'ID':id,'clube':clube,'cartao':'Vermelho','quantidade_cartoes':0,'mandante':1}

        #Incluindo a linha no Data Frame    
        df_cartoes_tratados_temp = df_cartoes_tratados_temp.append(nova_linha,ignore_index=True)   
```
