# Análise dos Dados de Cartões

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

os.chdir('D:\\Users\\Diogo\\Documents\\Pós Graduação\\Módulos\\TCC\\DataSets\\Futebol Brasileiro')
```

```python
#Importando o data set dos dados das partidas dos campeonatos brasileiros e os dados dos cartões amarelos

df_cartoes=pd.read_csv('df_cartoes.csv',sep=',')

df_faltas=pd.read_csv('df_faltas.csv',sep=',')
```

```python
#Renomeando a coluna partida_id para ID para o Data Frame de Faltas, para executar o merge
df_faltas.rename(columns={'partida_id':'ID'},inplace=True)
df_faltas.head()
```

```python
#Verificando a dimensão do Dataframe
df_cartoes.shape
```

```python
#Verificando a dimensão do Dataframe
df_faltas.shape
```

```python
#Verificando a dimensão do Dataframe
df_faltas.head()
```

```python
#Exetucando o Merge dos DataFrames 
df_analise = df_cartoes.merge(df_faltas, on='ID',how='right')
df_analise.head()
```

```python
#Verificando o os dados de colunas e registros do Dataframe
df_analise.info()
```

```python
#Existem registros faltantes no Data Frame de Cartões em comparação com o data frame de faltas. Ao executar o merge esses
#registros foram preenchidos como nulos no data frame resultado. Portanto, iremos preencher com zero esses dados nulos

df_analise.fillna(0,inplace=True)
df_analise.info()
```

```python
#Identificamos que as colunas numéricas do primeiro Data Frame foram mescladas com o tipo float. Vamos ajustar alterando para
#para o tipo inteiro

df_analise['qtd_cartao_amarelo_mandante']=df_analise['qtd_cartao_amarelo_mandante'].astype('int64')
df_analise['qtd_cartao_vermelho_mandante']=df_analise['qtd_cartao_vermelho_mandante'].astype('int64')
df_analise['qtd_cartao_amarelo_visitante']=df_analise['qtd_cartao_amarelo_visitante'].astype('int64')
df_analise['qtd_cartao_vermelho_visitante']=df_analise['qtd_cartao_vermelho_visitante'].astype('int64')
df_analise['total_cartoes_mandante']=df_analise['total_cartoes_mandante'].astype('int64')
df_analise['total_cartoes_visitante']=df_analise['total_cartoes_visitante'].astype('int64')
df_analise['total_cartoes_amarelos_x']=df_analise['total_cartoes_amarelos_x'].astype('int64')
df_analise['total_cartoes_vermelhos_x']=df_analise['total_cartoes_vermelhos_x'].astype('int64')
df_analise['total_cartoes_partida_x']=df_analise['total_cartoes_partida_x'].astype('int64')

df_analise.info()
```

```python
#Inserindo uma coluna para comparando os dados de cartoes amarelos para o mandante, verificando se as quantidades estão iguais
#para os dois data frames.

df_analise['status_analise_amarelo_mandante'] = df_analise.apply(lambda x: 
'Correto' if x['qtd_cartao_amarelo_mandante']==x['cartao_amarelo_mandante'] else 'Incorreto',axis=1 )
```

```python
#df_analise.drop('status_analise_amarelo',axis=1,inplace=True)
df_analise[['qtd_cartao_amarelo_mandante','cartao_amarelo_mandante','status_analise_amarelo_mandante']]
```

```python
#Inserindo uma coluna para comparar os dados de cartoes amarelos para o visitante, verificando se as quantidades estão iguais.

df_analise['status_analise_amarelo_visitante'] = df_analise.apply(lambda x: 
'Correto' if x['qtd_cartao_amarelo_visitante']==x['cartao_amarelo_visitante'] else 'Incorreto',axis=1 )
```

```python
#Conferindo as colunas
df_analise[['qtd_cartao_amarelo_visitante','cartao_amarelo_visitante','status_analise_amarelo_visitante']]
```

```python
#Inserindo uma coluna geral comparando os dados de cartoes amarelos, verificando os status.

df_analise['status_analise_amarelo'] = df_analise.apply(lambda x: 
'Correto' if (x['status_analise_amarelo_mandante']=='Correto') and (x['status_analise_amarelo_visitante']=='Correto') 
                                                        else 'Incorreto',axis=1 )
```

```python
#df_analise.info()
df_analise.query("(status_analise_amarelo_mandante=='Incorreto') and (status_analise_amarelo_visitante=='Incorreto')")
```

```python
#Agora iremos executar a análise dos cartões vermelhos
#Inserindo uma coluna para comparando os dados de cartoes vermelhos para o mandamte, verificando se as quantidades estão iguais.

df_analise['status_analise_vermelho_mandante'] = df_analise.apply(lambda x: 
'Correto' if x['qtd_cartao_vermelho_mandante']==x['cartao_vermelho_mandante'] else 'Incorreto',axis=1 )
```

```python
#Inserindo uma coluna para comparando os dados de cartoes vermelhos para o visitante, verificando se as quantidades estão iguais.

df_analise['status_analise_vermelho_visitante'] = df_analise.apply(lambda x: 
'Correto' if x['qtd_cartao_vermelho_visitante']==x['cartao_vermelho_visitante'] else 'Incorreto',axis=1 )
```

```python
#Inserindo uma coluna geral comparando os dados de cartoes vermelhos, verificando os status.

df_analise['status_analise_vermelho'] = df_analise.apply(lambda x: 
'Correto' if (x['status_analise_vermelho_mandante']=='Correto') and (x['status_analise_vermelho_visitante']=='Correto') else 'Incorreto',axis=1 )
```

```python
#Verificando os registros do Dataframe
df_analise.head()
```

```python
#Inserindo uma coluna geral comparando os status das análises de cartões vermelhos e amarelos.

df_analise['status_analise'] = df_analise.apply(lambda x: 
'Correto' if (x['status_analise_amarelo']=='Correto') and (x['status_analise_vermelho']=='Correto') else 'Incorreto',axis=1 )
```
