# 3.4 Compilação dos Dados

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

os.chdir('D:\\Users\\Diogo\\Documents\\Pós Graduação\\Módulos\\TCC\\DataSets\\Futebol Brasileiro')
```

```python
#Carregando os dados do Data Frame de cartões

df_cartoes=pd.read_csv('df_cartoes_tratados.csv',sep=',')
df_cartoes.head()
```

```python
#Verificando as dimensões do Dataframe
df_cartoes.shape
```

```python
#Carregando os dados do Data Frame de faltas

df_faltas=pd.read_csv('df_faltas.csv',sep=',')
df_faltas.head()
```

```python
#Verificando as dimensões do Dataframe
df_faltas.shape
```

```python
#Como identificamos anteriormente os dados relevantes no Data Frame de faltas se iniciam a partir do campeonato 13. Até o 
#campeonato 11 não temos praticamente os dados e no campeonato 12 temos os dados parciais. Portanto, iremos equalizar os dados
# de todos os Data Frames a partir desse recorte.

df_faltas=df_faltas[df_faltas.campeonato>12]
df_faltas.head()
```

```python
#Verificando as dimensões do Dataframe
df_faltas.shape
```

```python
#Filtrando o campeonato 12
df_cartoes=df_cartoes[df_cartoes.campeonato>12]
df_cartoes.head()
```

```python
#Verificando as dimensões do Dataframe
df_cartoes.shape
```

```python
# Iremos fazer a mescla dos dados a partir do Data Frame de cartões, tendo em vista que ele possui a maior quantidade de campos
# que serão utilizados no desenvolvimento do trabalho
```

# IDENTIFICANDO CAMPOS RELEVANTES NO DATA FRAME DE FALTAS


```python
#Verificar os registros e colunas do Dataframe
df_faltas.info()
```

```python
#Removendo os campos idênticos do data frame de faltas para não ocorrer duplicidade

df_faltas_tratado = df_faltas[['ID','passes_mandante','faltas_mandante','passes_visitante', 'faltas_visitante','total_faltas_partida',
                              'total_passes_partida']]
```

```python
#Realizando a mescla de dados dos Data Frame de cartões e faltas

df_dados=df_cartoes.merge(df_faltas_tratado,on='ID',how='outer')
df_dados.head()
```

```python
df_dados.shape
```

```python
df_dados.info()
```

```python
#Importando o Data frame dos dados de jogos

df_jogos=pd.read_csv('campeonato-brasileiro-full.csv',sep=',')
df_jogos.head()
```

```python
df_jogos.info()
```

```python
#Selecionando as colunas com os dados que serão mesclados

df_jogos_tratado = df_jogos[['ID','mandante_Placar','visitante_Placar', 'vencedor','rodata']]
df_jogos_tratado.info()
```

```python
# Carregando a mescla com alguns campos do Data Frame de jogos 
df_dados=df_dados.merge(df_jogos_tratado,on='ID',how='left')
df_dados.head()
```

```python
df_dados.shape
```

```python
#Iremos remover alguns campos para verificarmos a correlação entre os dados. Os dados serão removidos pois não acrescentam
# relevância como por exemplo o ID das partidas. São eles: ID, campeonato e todos os campos de totais de cartões. Os campos
# de totais serão removidos pois tem relação com os alvos e não serão inputados para o modelo de predição.

df_dados_corr = df_dados[['qtd_cartao_amarelo_mandante','qtd_cartao_vermelho_mandante','qtd_cartao_amarelo_visitante',
                         'qtd_cartao_vermelho_visitante','faltas_mandante','faltas_visitante',
                          'passes_mandante','passes_visitante','mandante_Placar','visitante_Placar',
                         'vencedor','rodata']]
```

```python
#Identificando a Matriz de correlação dos Dados
matriz_correlacao = df_dados_corr.corr()
matriz_correlacao
```

```python
#Verificando a correlacao para a variável de cartões amarelos para o mandante e visitante
df_amarelo_mandante_corr=matriz_correlacao['qtd_cartao_amarelo_mandante'].sort_values(ascending=False).reset_index()
df_amarelo_mandante_corr['qtd_cartao_amarelo_mandante']=df_amarelo_mandante_corr['qtd_cartao_amarelo_mandante'].round(2)
df_amarelo_mandante_corr
```

```python
#Separando a matriz de correlação dos dados de cartões amarelos
df_amarelo_visitante_corr=matriz_correlacao['qtd_cartao_amarelo_visitante'].sort_values(ascending=False).reset_index()
df_amarelo_visitante_corr['qtd_cartao_amarelo_visitante']=df_amarelo_visitante_corr['qtd_cartao_amarelo_visitante'].round(2)
df_amarelo_visitante_corr
```

```python
#Separando a matriz de correlação dos dados de cartões vermelhos
df_vermelho_mandante_corr=matriz_correlacao['qtd_cartao_vermelho_mandante'].sort_values(ascending=False).reset_index()
df_vermelho_mandante_corr['qtd_cartao_vermelho_mandante']=df_vermelho_mandante_corr['qtd_cartao_vermelho_mandante'].round(2)
df_vermelho_mandante_corr
```

```python
#Código para construir o gráfico para os cartões amarelos de mandantes

fig,ax=plt.subplots(figsize=(15,10))

bar_container=ax.barh(df_amarelo_mandante_corr['index'],df_amarelo_mandante_corr['qtd_cartao_amarelo_mandante'],align='center',color='#EEE8AA')

for pY, pX in enumerate(df_amarelo_mandante_corr['qtd_cartao_amarelo_mandante']):
    ax.annotate(pX, xy=(pX,pY))

plt.title('Correlações com a Quantidade de Cartões Amarelos - Mandante')
plt.xlabel('Correlação')
plt.ylabel('Variável')
plt.show()
```

```python
#Código para construir o gráfico para os cartões amarelos de visitantes

fig,ax=plt.subplots(figsize=(15,10))

bar_container=ax.barh(df_amarelo_visitante_corr['index'],df_amarelo_visitante_corr['qtd_cartao_amarelo_visitante'],align='center',color='#EEE8AA')

for pY, pX in enumerate(df_amarelo_visitante_corr['qtd_cartao_amarelo_visitante']):
    ax.annotate(pX, xy=(pX,pY))

plt.title('Correlações com a Quantidade de Cartões Amarelos - Visitante')
plt.xlabel('Correlação')
plt.ylabel('Variável')
plt.show()
```

```python
#Código para construir o gráfico para os cartões vermelhos para mandantes

fig,ax=plt.subplots(figsize=(15,10))

bar_container=ax.barh(df_vermelho_mandante_corr['index'],df_vermelho_mandante_corr['qtd_cartao_vermelho_mandante'],align='center',color='#EEE8AA')

for pY, pX in enumerate(df_vermelho_mandante_corr['qtd_cartao_vermelho_mandante']):
    ax.annotate(pX, xy=(pX,pY))

plt.title('Correlações com a Quantidade de Cartões Vermelhos - Mandante')
plt.xlabel('Correlação')
plt.ylabel('Variável')
plt.show()
```

```python
#Código para construir o gráfico para os cartões vermelhos para visitantes

fig,ax=plt.subplots(figsize=(15,10))

bar_container=ax.barh(df_vermelho_visitante_corr['index'],df_vermelho_visitante_corr['qtd_cartao_vermelho_visitante'],align='center',color='#EEE8AA')

for pY, pX in enumerate(df_vermelho_visitante_corr['qtd_cartao_vermelho_visitante']):
    ax.annotate(pX, xy=(pX,pY))

plt.title('Correlações com a Quantidade de Cartões Vermelhos - Visitante')
plt.xlabel('Correlação')
plt.ylabel('Variável')
plt.show()
```
