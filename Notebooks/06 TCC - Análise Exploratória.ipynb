```python
#Criando os Dataframes df_grafico_mandante e df_grafico_visitante

df_grafico_mandante=df_dados[['clube_mandante','faltas_mandante','qtd_cartao_amarelo_mandante','qtd_cartao_vermelho_mandante','total_cartoes_mandante']].groupby(['clube_mandante']).sum()
df_grafico_visitante=df_dados[['clube_visitante','faltas_visitante','qtd_cartao_amarelo_visitante','qtd_cartao_vermelho_visitante','total_cartoes_visitante']].groupby(['clube_visitante']).sum()
```

```python
#Verificando o Dataframe
df_grafico_mandante
```

```python
#Verificando o Dataframe
df_grafico_visitante
```

```python
#Refazendo os índices dos Dataframes para renomear as colunas
df_grafico_mandante.reset_index(inplace=True)
df_grafico_visitante.reset_index(inplace=True)

df_grafico_mandante.rename(columns={'clube_mandante':'clube'},inplace=True)
df_grafico_visitante.rename(columns={'clube_visitante':'clube'},inplace=True)
```

```python
#Mesclando os dados

df_grafico = df_grafico_mandante.merge(df_grafico_visitante, on='clube')
df_grafico.info()
```

```python
#Calculando as colunas de totais para o Dataframe

df_grafico['total_faltas']=df_grafico[['faltas_mandante','faltas_visitante']].sum(axis=1)
df_grafico['total_cartoes']=df_grafico[['total_cartoes_mandante','total_cartoes_visitante']].sum(axis=1)
df_grafico['total_cartao_amarelo']=df_grafico[['qtd_cartao_amarelo_mandante','qtd_cartao_amarelo_visitante']].sum(axis=1)
df_grafico['total_cartao_vermelho']=df_grafico[['qtd_cartao_vermelho_mandante','qtd_cartao_vermelho_visitante']].sum(axis=1)
df_grafico
```

```python
#Código para a construção dos dados

df_grafico.sort_values(by = ['total_faltas'],ascending=False,inplace=True)

fig,ax=plt.subplots(figsize=(20,15))
bar_container=ax.barh(df_grafico['clube'],df_grafico['total_faltas'],align='center',color='#EEE8AA')
bar_container=ax.barh(df_grafico['clube'],df_grafico['total_cartao_amarelo'],align='center',label='Quantidade de Cartões Amarelos',color='#FFD700')
bar_container=ax.barh(df_grafico['clube'],df_grafico['total_cartao_vermelho'],align='center',label='Quantidade de Cartões Vermelhos',color='#FF6347')
ax.set_yticks(df_grafico['clube'])
ax.invert_yaxis() 


ax.set(xlabel='Faltas', ylabel='Clube',
       title='Quantidade de faltas por clube Série A - 2015/2022')
ax.xaxis.grid(linestyle="--", linewidth=1)

for pY, pX in enumerate(df_grafico['total_faltas']):
    ax.annotate(pX, xy=(pX,pY))
    
for pY, pX in enumerate(df_grafico['total_cartao_amarelo']):
    ax.annotate(pX, xy=(pX,pY))
    
for pY, pX in enumerate(df_grafico['total_cartao_vermelho']):
    ax.annotate(pX, xy=(pX,pY))    
plt.legend()
plt.show()
```

```python
#Código para a construção do gráfico

df_grafico.sort_values(by = ['media_total_faltas'],ascending=False,inplace=True)

fig,ax=plt.subplots(figsize=(20,15))
bar_container=ax.barh(df_grafico['clube'],df_grafico['media_total_faltas'],align='center',color='#EEE8AA')
bar_container=ax.barh(df_grafico['clube'],df_grafico['media_total_cartao_amarelo'],align='center',label='Quantidade Média de Cartões Amarelos',color='#FFD700')
bar_container=ax.barh(df_grafico['clube'],df_grafico['media_total_cartao_vermelho'],align='center',label='Quantidade Média de Cartões Vermelhos',color='#FF6347')
ax.set_yticks(df_grafico['clube'])
ax.invert_yaxis()


ax.set(xlabel='Faltas', ylabel='Clube',
       title='Média de faltas por clube Série A - 2015/2022')
ax.xaxis.grid(linestyle="--", linewidth=1)

for pY, pX in enumerate(df_grafico['media_total_faltas']):
    ax.annotate(pX, xy=(pX,pY))
    
for pY, pX in enumerate(df_grafico['media_total_cartao_amarelo']):
    ax.annotate(pX, xy=(pX,pY))
    
for pY, pX in enumerate(df_grafico['media_total_cartao_vermelho']):
    ax.annotate(pX, xy=(pX,pY))    
plt.legend()
plt.show()
```
